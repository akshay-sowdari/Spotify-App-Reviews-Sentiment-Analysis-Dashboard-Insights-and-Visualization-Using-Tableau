# Adding Sentiment Column

```python
data['Sentiment']="YetToUpdate"
```

```python
data
```

```python
t =data['Review']
t
```

```python
print(type(t))
```

```python
cl=[]
for i in t:
    blob = TextBlob(i)
    cl.append(blob.sentiment.polarity)
    
```

```python
print(cl)
```

```python

for i in range(len(cl)):
    if cl[i] > 0:
        cl[i] = 'Positive'
    elif cl[i] == 0:
        cl[i] = 'Neutral'
    else:
        cl[i] = 'Negative'

print(cl)
```

```python
data['Sentiment']=cl
```
